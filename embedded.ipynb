```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### Sentences

sent=[
    "the glass of milk",
    "the glass of juice",
    "the cup of tea",
    "I am a good man",
    "I am a good developer",
    "Understand the meaning of words",
    "your videos are good"
]
```

```python
sent
```

```python
### Define the vocalbulary size
voc_size=10000
```

```python
### One Hot representation
one_hot_representation=[one_hot(words,voc_size) for words in sent]
```

```python
sent
```

```python
one_hot_representation
```

```python
### Word embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
### Notice how the sentences are of different lengths. 
### We have to convert them to the same length.
```

```python
### setting the sentences length. lets say, 8 for this example
sent_length=8
embedded_docs=pad_sequences(one_hot_representation,padding="pre",maxlen=sent_length)
### pad_sequences adds extra words to the sentences with "0" as vector to make all sentences of same length
### Used padding="pre" to add new words at the starting
```

```python
print(embedded_docs)
```

```python
### feature representation
dimension=10
```

```python
# ...existing code...
model = Sequential()
model.add(Embedding(voc_size, dimension, input_length=sent_length))
model.compile("adam", "mse")
model.build(input_shape=(None, sent_length))
# ...existing code...
```

```python

model.summary()
```

```python
model.predict(embedded_docs)
```
